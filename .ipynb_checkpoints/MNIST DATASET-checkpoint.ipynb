## traing slicing and testing slicing

```python
x_train=x[0:6000]
```

```python
x_test=x[6000:70000]
```

```python
y_train,y_test=x[0:6000],x[6000:70000]
```

## suffeling the traing set

```python
import numpy as np
suffel_index=np.random.permutation(6000)
x_train,y_train=x_train[suffel_index],y_train[suffel_index]
```

## creating a 2 detector

```python
y_train=y_train.astype(np.int8)
y_test=y_test.astype(np.int8)
y_train_2=(y_train==2)
y_test_2=(y_test==2)
```

```python
y_test_2
y_train
```

```python
y_train_2
```

## classification

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(tol=0.1,solver="lbfgs")
```

```python
clf.fit(x_train, y_train_2)
```

```python
clf.predict([some_digit])
```

## cross validation

```python
from sklearn.model_selection import cross_val_score
a=cross_val_score(clf,x_train,y_train_2,cv=3,scoring="accuracy")
```
